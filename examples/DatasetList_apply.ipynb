```python
import pandas as pd
from caits.dataset import DatasetList, CoreArray
from caits.loading import csv_loader
import numpy as np
```

```python
init_data = csv_loader("data/GestureSet_small")

print(init_data)
```

```python
tmp = [
    (init_data["X"][i], init_data["y"][i], init_data["id"][i])
    for i in range(len(init_data["X"]))
]

X, y, _id = zip(*tmp)
```

```python
data = DatasetList(
    [CoreArray(x.values, axis_names={"axis_1": x.columns}) for x in X],
    CoreArray(np.array(y)),
    _id
)
```

```python
from caits.filtering import filter_butterworth

tr_data = data.apply(
    filter_butterworth,
    fs = 1000,
    cutoff_freq = 5,
    axis_names_X={"axis_1": data.get_axis_names_X()["axis_1"]},
)
```
