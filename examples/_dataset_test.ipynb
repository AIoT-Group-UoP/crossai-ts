```python
data = csv_loader_regression("../examples/data/AirQuality.csv", sep=";", decimal=",")
data

X = [{}]
y = [{}]

for df in data["X"]:
    X[0][df.columns[0]] = df.values

# for df in data["y"]:
#     X[0][df.columns[0]] = df.values

for df in data["y"]:
    y[0][df.columns[0]] = df.values

X, y
```

```python
normalize_tr = FunctionTransformer(normalize_signal)
butterworth_tr = FunctionTransformer(filter_butterworth, fs=4, filter_type="lowpass", cutoff_freq=0.5)

tmp = normalize_tr.fit_transform(dataset)
```

```python
inst1 = dataset.X[0]
cols1 = list(inst1.keys())
vals1 = np.stack([v[:, 0] for v in inst1.values()], axis=1)
df1 = pd.DataFrame(vals1, columns=cols1)
df1
```

```python
inst2 = tmp.X[0]
cols2 = list(inst2.keys())
vals2 = np.stack([v[:, 0] for v in inst2.values()], axis=1)
df2 = pd.DataFrame(vals2, columns=cols2)
df2
```

```python
fig = plot_signal(df1.values)
```
