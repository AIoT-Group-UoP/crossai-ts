# HPSS on Uni-Axis data test

In this notebook, we check that HPSS functions work as expected on Multi-Axis data.

# Test uni-axis

```python
from IPython.display import Audio
from caits.loading import wav_loader
from caits.visualization import plot_signal
from caits import hpss
```

```python
from caits.loading import wav_specs_check
```

```python
audio_params = wav_specs_check("data/yes.wav")#%%
audio_params
```

```python
audio_mono, sample_rate = wav_loader("data/yes.wav")

print(f"shape: {audio_mono.shape} - Sampling rate: {sample_rate}")
Audio(audio_mono.values.T, rate=sample_rate)
```

```python
audio_arr = audio_mono.values
```

```python
plot_signal(
    audio_arr,
    sr=sample_rate,
    return_mode=False
)
```

# Perform HPSS

```python
hpss_audio = hpss.hpss(audio_arr)
hpss_audio[0].shape, hpss_audio[1].shape
```

```python
plot_signal(
    hpss_audio[0],
    sr=sample_rate,
    return_mode=False
)
```

```python
plot_signal(
    hpss_audio[1],
    sr=sample_rate,
    return_mode=False
)
```
