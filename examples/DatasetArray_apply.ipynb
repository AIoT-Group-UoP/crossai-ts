```python
import pandas as pd
from caits.dataset import CoreArray, DatasetArray
```

```python
init_data = pd.read_csv("data/AirQuality.csv", sep=";", decimal=",")
print(init_data)
```

```python
init_data["Datetime"] = init_data["Date"] + " " + init_data["Time"]
init_data = init_data.drop(columns=["Date", "Time"]).set_index("Datetime")
init_data = init_data[init_data.index.isna() == False].reset_index()
print(init_data)
```

```python
X_df = init_data.iloc[:, :-5]
X_values = X_df.values
X_axis_names = {
    "axis_0": X_df.index.values,
    "axis_1": X_df.columns
}

X = CoreArray(X_values, X_axis_names)
print(X)
```

```python
y_df = init_data.iloc[:, -5:-2]
y_values = y_df.values
y_axis_names = {
    "axis_0": y_df.index.values,
    "axis_1": y_df.columns
}

y = CoreArray(y_values, y_axis_names)
print(y)
```

```python
data = DatasetArray(X, y)
print(data)
```
